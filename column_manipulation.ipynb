```python
import pandas as pd
```

```python
aq_df = pd.read_csv("Airquality.csv", index_col=0, parse_dates=True)
```

```python
aq_df
```

### want to express the NO2 concentration of the station in London in mg/m3

```python
aq_df["london_mg_per_cub_m"] = aq_df["station_london"] * 1.882
```

```python
aq_df
```

### want to check the ratio of the values in Paris versus antewerp and save it new column

```python
aq_df["ration_paris_antwerp"] = ((aq_df["station_paris"]) / aq_df["station_antwerp"])
```

```python
aq_df
```

```python
aq_df["percentage_paris_antwerp"] = aq_df["ration_paris_antwerp"] * 100
```

```python
aq_df
```

### want to rename the data columns to the corresponding station identifiers

```python
aq_df_renamed = aq_df.rename(columns={"station_antwerp":"antewerp", "station_paris": "paris", "station_london": "London"})
```
